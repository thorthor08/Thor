````python
import pandas as pd
import seaborn as sns
import numpy as np
from ipywidgets import interactive
import matplotlib.pyplot as plt

def read_csv_from_googledrive(url):
  if(url.startswith("https://drive.google.com/uc?id=")):
    return pd.read_csv(url)
  url='https://drive.google.com/uc?id=' + url.split('/')[-2]
  df=pd.read_csv(url)
  return df

df = read_csv_from_googledrive("https://drive.google.com/file/d/1CPRMSRu7n2zkJjEMRMoHKuJQB8BWDwzc/view")
display(df)
df_copy = df.copy()
````

````python
df_copy.isna().sum()
df_clean = df_copy.copy()
display(df_clean.isna().sum())
df_clean.ffill(inplace=True)
display(df_clean.isna().sum())
df_clean
````

<div dir='rtl'>

**2)**

זיהוי וטיפול בערכים חריגים:

````python
q1, q3 = np.percentile(df_clean['Age'], [25, 75])
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
for val in df_clean['Age']:
  if val < lower_bound or val > upper_bound:
    df_clean['Age'] = df_clean['Age'].replace(val, df_clean['Age'].mean())
df_clean

````

בצעו מחקר של משתנה בודד כפי שלמדנו:

````python
def plot_categorical_variable(column):
    plt.figure(figsize=(10, 6))
    sns.histplot(x=column, data=df_clean)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

display(interactive(plot_categorical_variable, column=df_clean.select_dtypes(include=['object']).columns))


def plot_numerical_variable(column, bins):
    plt.figure(figsize=(10, 6))
    sns.histplot(x=column, data=df_clean, bins=bins)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

display(interactive(plot_numerical_variable, column=df_clean.select_dtypes(include=['number']).columns, bins=(1, 44)))
````
