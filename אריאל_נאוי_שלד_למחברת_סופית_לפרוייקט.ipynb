````python
import pandas as pd
import seaborn as sns
import numpy as np
from ipywidgets import interactive
import matplotlib.pyplot as plt


def read_csv_from_googledrive(url):
  if(url.startswith("https://drive.google.com/uc?id=")):
    return pd.read_csv(url)
  url='https://drive.google.com/uc?id=' + url.split('/')[-2]
  df=pd.read_csv(url)
  return df

df = read_csv_from_googledrive("https://drive.google.com/file/d/1CPRMSRu7n2zkJjEMRMoHKuJQB8BWDwzc/view")
display(df)
df_copy = df.copy()
````

##חקירת נתונים ראשונית:

**1)**

טיפול בערכים חסרים:

````python
df_copy.isna().sum()
df_clean = df_copy.copy()
display(df_clean.isna().sum())
df_clean.ffill(inplace=True)
display(df_clean.isna().sum())
df_clean
````

<div dir='rtl'>

**2)**

זיהוי וטיפול בערכים חריגים:

````python
q1, q3 = np.percentile(df_clean['Age'], [25, 75])
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
for val in df_clean['Age']:
  if val < lower_bound or val > upper_bound:
    df_clean['Age'] = df_clean['Age'].replace(val, df_clean['Age'].mean())
df_clean

````

````python
def plot_categorical_variable(column):
    plt.figure(figsize=(7, 3))
    sns.histplot(x=column, data=df_clean)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

display(interactive(plot_categorical_variable, column=df_clean.select_dtypes(include=['object']).columns))


def plot_numerical_variable(column, bins):
    plt.figure(figsize=(7, 3))
    sns.histplot(x=column, data=df_clean, bins=bins)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()

display(interactive(plot_numerical_variable, column=df_clean.select_dtypes(include=['number']).columns, bins=(1, 44)))
````

#חלק 2: הכנה לקראת המודל:

<div dir='rtl'>

**2)**

התמודדות עם ערכים קטגוריאלים בינאריים:

````python
new_df['CryoSleep'] = new_df['CryoSleep'].map({True: 1, False: 0})
new_df
````

<div dir='rtl'>

**3)**

התמודדות עם ערכים קטגוריאלים עם יותר מ-2 ערכים:

**טיפ:** חיזרו למצגת של שיעור 27, והסתכלו על ה-labelEncoder שראינו במהלך השיעור (שקף 48)

````python
Categorical_columns = new_df.select_dtypes(include=['object']).columns
count_dict = {}
for column in Categorical_columns:
    unique_values = new_df[column].unique()
    for value in unique_values:
        if value not in count_dict:
            count_dict[value] = len(count_dict)
    new_df[column] = new_df[column].map(count_dict)
    count_dict = {}

new_df
````

<div dir='rtl'>

**4)**

נרמול הנתונים:

````python
from sklearn.preprocessing import StandardScaler, MinMaxScaler

new_df_copy = new_df.copy()
num_columns = new_df_copy.select_dtypes(include=['number']).columns
num_columns = new_df_copy[num_columns]
min_max_scaler = MinMaxScaler()


final_train = pd.DataFrame(min_max_scaler.fit_transform(num_columns), columns=num_columns.columns)
final_train
````

<div dir='rtl'>

**6)**

סטנדרטיזציה של הנתונים:

````python
standard_scaler = StandardScaler()
final_train = pd.DataFrame(standard_scaler.fit_transform(num_columns), columns=num_columns.columns)
final_train = final_train.join(new_df_copy['Transported'])
final_train
````

<div dir='rtl'>

**7)**

סידור סט המבחן:

````python
test_set = read_csv_from_googledrive("https://drive.google.com/file/d/1r9og3oqXkLroMkIyG40gOUQkawLEbRd4/view")
display(test_set)
test_set_copy = test_set.copy()
train_set = final_train.copy()
test_set_copy.isna().any()
````

````python
test_set_copy['CryoSleep'] = test_set_copy['CryoSleep'].map({True: 1, False: 0})

Categorical_columns = test_set_copy.select_dtypes(include=['object']).columns
count_dict = {}
for column in Categorical_columns:
    unique_values = test_set_copy[column].unique()
    for value in unique_values:
        if value not in count_dict:
            count_dict[value] = len(count_dict)
    test_set_copy[column] = test_set_copy[column].map(count_dict)
    count_dict = {}

test_cop = test_set_copy.copy()[['HomePlanet', 'Gender', 'Age', 'CryoSleep', 'Deck', 'RoomNum', 'Transported']]


num_columns1 = test_cop.select_dtypes(include=['number']).columns
num_columns1 = test_cop[num_columns1]

final_test = pd.DataFrame(min_max_scaler.fit_transform(num_columns1), columns=num_columns1.columns)

final_test = pd.DataFrame(standard_scaler.fit_transform(num_columns1), columns=num_columns1.columns)
final_test = final_test.join(test_cop['Transported'])

test_set = final_test.copy()
test_set
````

<div dir='rtl'>

#חלק 3: אימון המודל:

<div dir='rtl'>

**1)**

מודל מוסתר:

````python
#@title הריצו את קטע הקוד
from sklearn.tree import DecisionTreeClassifier

def read_csv_from_googledrive(url):
  if(url.startswith("https://drive.google.com/uc?id=")):
    return pd.read_csv(url)
  url='https://drive.google.com/uc?id=' + url.split('/')[-2]
  df=pd.read_csv(url)
  return df

clf = DecisionTreeClassifier()
x = train_set.drop('Transported', axis=1)
y = train_set['Transported']
clf.fit(x, y)

y_test = test_set['Transported']
train_pred = clf.predict(x)
x_test = test_set.drop('Transported', axis=1)
test_pred = clf.predict(x_test)
````

````python
from sklearn.metrics import accuracy_score
acc_df = pd.DataFrame()
acc_df['Set'] = ['Train', 'Test']
acc_df['Accuracy'] = [accuracy_score(y, train_pred)*100, accuracy_score(y_test, test_pred)*100]
display(acc_df)
sns.barplot(data=acc_df, x='Set', y='Accuracy')
plt.title("Accuracy as a Function of Set")
plt.xlabel("Set")
plt.ylabel("Accuracy")
plt.show()
````

<div dir='rtl'>

**3)**

אימון מודלים נוספים:

````python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neighbors import NearestCentroid

model1 = KNeighborsClassifier(n_neighbors=3)
model1.fit(x, y)
k_pred = model1.predict(x_test)


model2 = NearestCentroid()
model2.fit(x, y)
n_pred = model2.predict(x_test)
acc = pd.DataFrame()
acc['Model'] = ['KNN', 'NCC']
acc['Accuracy'] = [accuracy_score(y_test, k_pred)*100, accuracy_score(y_test, n_pred)*100]
display(acc)

sns.barplot(data=acc, x='Model', y='Accuracy')
plt.title("Accuracy as a Function of Model")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.show()
````
